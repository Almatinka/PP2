```python
from datetime import datetime, timedelta

current_date = datetime.now()
new_date = current_date - timedelta(days=5)

print("Current Date:", current_date.strftime("%Y-%m-%d"))
print("Date 5 days ago:", new_date.strftime("%Y-%m-%d"))
```

```python
today = datetime.now()
yesterday = today - timedelta(days=1)
tomorrow = today + timedelta(days=1)

print("Yesterday:", yesterday.strftime("%Y-%m-%d"))
print("Today:", today.strftime("%Y-%m-%d"))
print("Tomorrow:", tomorrow.strftime("%Y-%m-%d"))
```

```python
def square_generator(N):
    for i in range(N + 1):
        yield i ** 2

for num in square_generator(10):
    print(num, end=" ")
```

```python
def even_numbers(n):
    for i in range(0, n + 1, 2):
        yield i

print(", ".join(map(str, even_numbers(20))))
```

```python
def divisible_by_3_and_4(n):
    for i in range(n + 1):
        if i % 3 == 0 and i % 4 == 0:
            yield i

print(list(divisible_by_3_and_4(50)))
```

```python
def squares(a, b):
    for i in range(a, b + 1):
        yield i ** 2

for value in squares(3, 7):
    print(value)
```

```python
def countdown(n):
    while n >= 0:
        yield n
        n -= 1

for num in countdown(10):
    print(num, end=" ")
```

```python
num_sides = 4
side_length = 25

polygon_area = (num_sides * side_length ** 2) / (4 * math.tan(math.pi / num_sides))
print(f"The area of the polygon is: {polygon_area:.2f}")
```
